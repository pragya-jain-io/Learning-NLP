```python
paragraph = '''As we all know, a paragraph is a group of sentences that are connected and make absolute sense. While writing a long essay or letter, we break them into paragraphs for better understanding and to make a well-structured writing piece. Paragraph writing on any topic is not only about expressing your thoughts on the given topic, but it is also about framing ideas about the topic and making it convenient for the readers to follow it. In English paragraph writing, it is essential to focus on the writing style, i.e., the flow and connection between the sentences. 
Therefore, a paragraph must be written in simple language in order to avoid any interruption while reading.'''
```

Tokenization<br>


```python
from nltk.tokenize import word_tokenize
```

```python
paragraph_tokens = word_tokenize(paragraph)
paragraph_tokens[:10]
```

```python
len(paragraph_tokens)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
for word in paragraph_tokens:
    fdist[word.lower()]+=1
fdist
```

```python
fdist[',']
```

```python
len(fdist)
```

```python
fdist_top10 = fdist.most_common(10)
fdist_top10
```

```python
from nltk.tokenize import blankline_tokenize
paragraph_blank = blankline_tokenize(paragraph)
len(paragraph_blank)
```

Named entity recognition<br>



```python
from nltk import ne_chunk
ne_sent = "jane is driving to Philadelphia"
ne_tokens = word_tokenize(ne_sent)
ne_tags = nltk.pos_tag(ne_tokens)
ne_ner = ne_chunk(ne_tags)
print(ne_ner)
```

Chunking
- The chunking technique is used to break down a large piece of information into smaller, more manageable

 

```python
new_sentence = "A quick brown fox jumped over a lazy dog."
new_tokens = word_tokenize(new_sentence)
new_pos = nltk.pos_tag(new_tokens)
new_pos
```

```python
grammar_np = r"NP: {<DT>?<JJ>*<NN>}"
chunk_parser = nltk.RegexpParser(grammar_np)
result = chunk_parser.parse(new_pos)
result
```
